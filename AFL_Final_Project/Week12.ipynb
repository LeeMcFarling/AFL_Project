# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

#### **Answer**

So for the purposes of naming here, we will refer to top graph as 'plot A' and the bottom plot as 'plot B'. 

Advantages of Plot A: 

Here the main advantage to plotting in this way would be to be able to compare the structure and shape of two different data series, expecially if they have a different scale (if one series is a factor of 10 greater than the other for example), this way, we can talk about the relative behavior of the two series without having to add a potentially cumbersome second axis to the plot. 

Disadvantages of Plot A: 

A key disadvantage is that comparing values on the same day across the two weeks becomes more difficult. For instance, comparing the number of visitors on Day 6 requires looking back and forth between the two axes, mentally adjusting for scale, and visually aligning the data points. This makes day-by-day comparisons more cumbersome. In cases where the series are relatively close in magnitude  (like in this example) it would be more effective to plot them on the same axis to allow for direct visual comparison.


Advantages of Plot B: 

Because the scale of the two series in this case are very similar (both range from 70 to 130 visitors a day), it is very advantagous to plot the two of these on the same axis. This easily reveals patterns that are difficult to discern in plot A. (for example, series one peaks earlier on in the week -- at day 2 while the second series peaks at day 4.). It also simplifies direct comparisons, allowing us to easily see which week had more visitors on any given day.


Disadvantages of B: 

If the two series differed greatly in scale — for example, if one series ranged from 100 to 1,200 while the other ranged from 70 to 120 — plotting them on the same axis would be problematic. The larger range would dominate the space, making it hard to discern meaningful patterns in the smaller-scale series. In such cases, details in the lower range may be effectively flattened out (or if we did the opposite and kept the smaller scale, the larger series would be obscured.)

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```
